```python
import cv2
import numpy as np
import pyautogui
from cvzone.HandTrackingModule import HandDetector
```

```python
# finger clicker and cursor

cap = cv2.VideoCapture(0)
cap.set(3, 1280)
cap.set(4, 720)
detector = HandDetector(detectionCon=0.5, maxHands=1)
prevFingerCount = 0
pyautogui.FAILSAFE = False

while True:
    try:
        # Get image frame
        success, frame = cap.read()
        img = cv2.resize(frame, (pyautogui.size()[0], pyautogui.size()[1]))

        # Find the hand and its landmarks
        hands, img = detector.findHands(img)

        if hands:  # If hands are detected
            for hand in hands:
                lmList = hand["lmList"]  # Retrieve hand landmarks
                bbox = hand["bbox"]  # Retrieve bounding box coordinates

#                 x, y = lmList[8][1:]  # Coordinates of the tip of the index finger
#                 screen_x = np.interp(x, [0, bbox[2]], [0, pyautogui.size()[0]])
#                 screen_y = np.interp(y, [0, bbox[3]], [0, pyautogui.size()[1]])

                x = bbox[0]
                y = bbox[1]

                if lmList:
                    # Find how many fingers are up
                    fingers = detector.fingersUp(hand)
                    totalFingers = fingers.count(1)

                    if totalFingers in [0, 1, 2, 3, 4]:
                        screen_width, _ = pyautogui.size()
                        pyautogui.moveTo(screen_width - x, y, duration=pyautogui.MINIMUM_DURATION)
                    elif totalFingers == 5:
                        pyautogui.click()

                    cv2.putText(img, f'Fingers:{totalFingers}', (bbox[0] + 100, bbox[1] - 30),
                                cv2.FONT_HERSHEY_PLAIN, 2, (0, 255, 0), 2)
                    if totalFingers != prevFingerCount:
                        print(str(totalFingers) + " fingers held up")
                        prevFingerCount = totalFingers

        # Display
        cv2.imshow("Image", img)
        if cv2.waitKey(1) == ord('q'):  # Close window when 'q' key is pressed
            break

    except pyautogui.FailSafeException as e:
        print("Fail-safe exception occurred:", str(e))
        # Move cursor to the center of the screen
        screen_width, screen_height = pyautogui.size()
        center_x = screen_width // 2
        center_y = screen_height // 2
        pyautogui.moveTo(center_x, center_y)

# Release the video capture and close the window
cap.release()
cv2.destroyAllWindows()
```
